```python
r = 5
r1 = 3

c = symbols('c0:125')
```

```python
H = 0
for n in range(1, r):
    h = (-1/A) * c[n] * (A*exp(-A))**n
    H += h

len(Poly(H, exp(-A)).all_coeffs())
```

```python
g = 1
for n in range(1, r):
    F = 0
    for m in range(0, r1):
        f1 = (-1)**m * (c[n]*(A*exp(-A))**n)**m
        f2 = f1/factorial(m)
        F += f2
#     display(F)
    g = g*F
    
G = g.expand().collect(A*exp(-A))

len(Poly(G, exp(-A)).all_coeffs())
display(G)
```

```python
sol1 = exp(-A)*G + H
sol2 = sol1.expand().collect(A*exp(-A))
sol2
```

```python
coeff = Poly(sol2, exp(-A)).all_coeffs()
for i in range(0, len(coeff)):
    coeff[i] = coeff[i].subs({A:1})
    
coeff = coeff[::-1]
coeff
```
